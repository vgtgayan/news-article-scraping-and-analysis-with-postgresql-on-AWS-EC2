```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
def scrape_website(url: str):
    """
        Scrape contents of a given website(url)
        Params:
        url - String -- url of the website
    """
    result = requests.get(url)
    doc = BeautifulSoup(result.text, "html.parser")
    return doc
```

```python
# Test scrape_website()
url_ = "https://edition.cnn.com/"
doc_ = scrape_website(url_)
print(doc_.prettify())
```

```python
def get_title(tag: BeautifulSoup):
    """
        Returns title of the given tag
    """
    result = tag.find("h1", class_="pg-headline")
    return result.string
```

```python
def get_paragraphs(tag: BeautifulSoup):
    """
        Returns list of paragraphs inside the given tag
    """
    paragraphs = tag.find_all("div", class_="zn-body__paragraph")
    paragraphs_list = [p.string for p in paragraphs]
    return paragraphs_list
```

```python
def get_author(tag: BeautifulSoup):
    """Return author of an article"""
    author_line = tag.find("span", class_="metadata__byline__author")
    author_name = author_line.find("a")
    return author_name.string
```

```python
def get_posted_date(tag: BeautifulSoup):
    """Return posted date of an article"""
    time_tag = tag.find("p", class_="update-time")
    time_str = time_tag.contents[0]
    date_ = re.search(r'(\w+ \d+, \d{4})', time_str)
    return date_.group(0)
```

```python
get_posted_date(doc_3)
```
